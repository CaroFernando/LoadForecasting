```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# read dataframe

df = pd.read_csv('time_series_30min_singleindex.csv')
df['utc_timestamp'] = pd.to_datetime(df['utc_timestamp'])
time_load_df = df[['utc_timestamp', 'CY_load_actual_entsoe_transparency']]
```

```python
time_load_df = time_load_df.dropna()
time_load_df.reset_index(drop=True, inplace=True)
time_load_df.head(10)
```

```python
px.line(time_load_df, x='utc_timestamp', y='CY_load_actual_entsoe_transparency', title='Load in Cyprus')
```

```python
# load of jun 2018
mask = (time_load_df['utc_timestamp'] > '2018-06-01') & (time_load_df['utc_timestamp'] < '2018-07-01')

jundf = time_load_df[mask]
```
